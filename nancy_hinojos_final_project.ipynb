```python
#class used to return affordable, high quality wines according to user input
class affordable1:
    def __init__(self, variety):
        self.variety = variety
        
    def screen_output(self):
        selection = affordable.loc[affordable.variety == self.variety,['variety','price','winery','country']]
        print(selection)
```

## Interested in finding high rated wines at a more afforable price? Enter your selection below to return a list.

```python
#calls class according to user input 
#code lists out the wines in the affordable, high rated list for final user

quality = input("Are you interested in Syrah, Chardonnay, Pinot Noir? Type one of the options to see a select list of wines")
if quality not in {"Syrah","Chardonnay", "Pinot Noir"}:
    print("We were unable to process your request. Please try again.")

while quality not in {"Syrah","Chardonnay", "Pinot Noir"}:
    try:
        quality = input("Are you interested in Syrah, Chardonnay, Pinot Noir? Type one of the options to see a select list of wines")
        selection = affordable1(quality)
        
    except ValueError as err:
        print("That was not good input, please try again!")
        print("Error was", err)
print("\n Please see below for wine list with wines rated 99 or 100 and less than or equal to $90.")
selection.screen_output()
```
